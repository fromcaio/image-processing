## EXERCÍCIO 01
Carregar a imagem lena (images/lenaCor.bmp), convertê-la para monocromática, e aplicar a ela um filtro passa-alta obtido através da criação de uma máscara hp.

Hp = [-1 -1 -1, -1 8 -1, -1 -1 -1]

J1 = filter2(h1,I);

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Carregar a imagem
img = cv2.imread('../images/lenaCor.bmp')

# Converter para monocromática (escala de cinza)
if img is None:
    raise FileNotFoundError("Imagem não encontrada. Verifique o caminho.")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Definir o filtro passa-alta
hp_filter = np.array([[-1, -1, -1],
                      [-1,  8, -1],
                      [-1, -1, -1]], dtype=np.float32)

# Aplicar o filtro usando cv2.filter2D (equivalente ao filter2 do Octave/MATLAB)
# o parâmetro -1 indica que a profundidade (numero de bits de cada pixel) da imagem de saída será a mesma da imagem de entrada
# 0 poderia ser usado para converter a imagem para 8 bits mas ela já está em 8 bits
filtered_img = cv2.filter2D(img_gray, -1, hp_filter)

# Exibir as imagens
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original (Colorida)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem Monocromática')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_img, cmap='gray')
plt.title('Filtro Passa-Alta Aplicado')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## EXERCÍCIO 02
Carregar a imagem lena, convertê-la para monocromática e aplicar a ela um filtro de Robert obtido através da criação de uma máscara Robert. 

robert=[1 0, 0 -1] 

J2 = filter2(h2,Robert);

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Carregar a imagem
img = cv2.imread('../images/lenaCor.bmp')

# Converter para monocromática (escala de cinza)
if img is None:
    raise FileNotFoundError("Imagem não encontrada. Verifique o caminho.")

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Definir o filtro roberts
hp_filter1 = np.array([[1, 0],
                      [0, -1]], dtype=np.float32)
hp_filter2 = np.array([[0, 1],
                        [-1, 0]], dtype=np.float32)

# Aplicar o filtro usando cv2.filter2D (equivalente ao filter2 do Octave/MATLAB)
# o parâmetro -1 indica que a profundidade (numero de bits de cada pixel) da imagem de saída será a mesma da imagem de entrada
# 0 poderia ser usado para converter a imagem para 8 bits mas ela já está em 8 bits
filtered_img1 = cv2.filter2D(img_gray, -1, hp_filter1)
filtered_img2 = cv2.filter2D(img_gray, -1, hp_filter2)

filtered_img = filtered_img1 + filtered_img2

# Exibir as imagens
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Imagem Original (Colorida)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Imagem Monocromática')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(filtered_img, cmap='gray')
plt.title('Filtro de Roberts Aplicado')
plt.axis('off')

plt.tight_layout()
plt.show()
```
